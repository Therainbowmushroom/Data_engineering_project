```python
import pandas as pd
```

```python
file_id = "1zmNAMYzerjdrCZs51MmoWiHfvog-xdi3" 
file_url = f"https://drive.google.com/uc?id={file_id}"
df_lung = pd.read_csv(file_url)
df_lung.head(10)
```

```python
columns = df_lung.columns
for col in columns:
    print(df_lung[col].info())
```

Таким образом, явных категориальных данных в таблице два столбца: GENDER и LUNG_CANCER, но и остальные стобцы тоже содержат категориальные данные, кроме стобца AGE.

```python
df_lung.columns
```

Видно, что названия колонок содержат ненужные пробелы, написаны капсом, пробелы между словами нужно заменить на нижнее подчёркивание.

```python
columns = df_lung.columns
dict_rename = {}
for col in columns:
    dict_rename[col] = col.lower().rstrip().replace(' ', '_')
df_lung.rename(columns = dict_rename, inplace=True)
df_lung.columns
```

Успешно привели названия столбцов к нормативному.

```python
df_lung.isnull().sum() #пропусков в столбцах нет
```

Замена значений в столбцах: 'smoking', 'yellow_fingers', 'anxiety','peer_pressure', 'chronic_disease', 'fatigue', 'allergy', 'wheezing', 'alcohol_consuming', 'coughing', 'shortness_of_breath','swallowing_difficulty', 'chest_pain' - на значения True и False и приведение значений к типу bool.

```python
column_age_lung = ['smoking', 'yellow_fingers', 'anxiety',
       'peer_pressure', 'chronic_disease', 'fatigue', 'allergy', 'wheezing',
       'alcohol_consuming', 'coughing', 'shortness_of_breath',
       'swallowing_difficulty', 'chest_pain']
for column in column_age_lung:    
    df_lung[column] = df_lung[column].map({2:True, 1:False}).astype(bool)
df_lung.info()
```

Замена значений в столбце lung_cancer с YES и NO на True и False соотвественно. И замена значений в gender с M и F на 1 и 0 соотвественно.

```python
df_lung['lung_cancer'] = df_lung['lung_cancer'].map({'YES':True, 'NO':False}).astype(bool)
df_lung['gender'] = df_lung['gender'].map({'M':1, 'F':0}).astype('category')
df_lung.info()
```
